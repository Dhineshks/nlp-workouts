```python
import pandas as pd

train_data = pd.read_csv("Corona_NLP_train.csv",encoding='latin-1')
train_data.head()
```

```python
# lowercasing text

train_data['OriginalTweet'] = train_data['OriginalTweet'].astype(str).str.lower()
train_data.head(6)
```

```python
# A RegexpTokenizer splits a string into substrings using regular expressions. \w+ matches Unicode word characters with one or more occurrences.

from nltk.tokenize import RegexpTokenizer

regexp = RegexpTokenizer('\w+')

train_data['tweet'] = train_data['OriginalTweet'].apply(regexp.tokenize)
train_data.head(6)
```

```python
import nltk
nltk.download('stopwords')
```

```python
# stopwords are some common words like "the", "and", "I", "is", "are" etc, and these words don't provide any useful information. Hence these words can be dropped.

from nltk.corpus import stopwords

stop_words = stopwords.words('english')

custom_stopwords = ['https']
stop_words.extend(custom_stopwords)

train_data['tweet'] = train_data['tweet'].apply(lambda x: [item for item in x if item not in stop_words])
train_data.head(6)
```

```python
# nltk.download('wordnet')
# nltk.download('omw-1.4')
```

```python
# stemming is the process of removing word affixes. For example the word "Jumping" can be reduced to "Jump".

from nltk.stem.porter import PorterStemmer 

def apply_stemming(tokenized_column):

  stemmer = PorterStemmer()
  return [stemmer.stem(word) for word in tokenized_column]  
```

```python
train_data['tweet'] = train_data.apply(lambda x: apply_stemming(x['tweet']), axis=1)

train_data.head(7)
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
# lemmatization is the process that basically reduces each word to it's root. For example the word "funny" can be reduced to "fun".
# POS tag assign part of speech to every word in a string. Using POS can improve the result of lemmatization.

from nltk.corpus import wordnet as wn
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from collections import defaultdict

def apply_lemmatizer(stemmed_column):
  tag_map = defaultdict(lambda : wn.NOUN)
  tag_map['J'] = wn.ADJ
  tag_map['V'] = wn.VERB
  tag_map['R'] = wn.ADV

  wordnet_lem = WordNetLemmatizer()
  return [wordnet_lem.lemmatize(word, tag_map[tag[0]]) for word, tag in pos_tag(stemmed_column)]
```

```python
train_data['tweet'] = train_data.apply(lambda x: apply_lemmatizer(x['tweet']), axis=1)
train_data.head(7)
```

```python
def rejoin_words(col):

  return (" ".join(col))
```

```python
train_data['tweet'] = train_data.apply(lambda x: rejoin_words(x['tweet']), axis=1)
train_data.head(7)
```
